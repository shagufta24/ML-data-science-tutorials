```python
# Function to plot the cumulative cases for a country over the months
def plot_cumulative_cases(country_name):
    country = who_time_series[who_time_series['Country'] == country_name]
    plt.plot(country['Date_reported'], country['Cumulative_cases'])
    plt.title('{}: Cumulative Reported Cases'.format(country_name))
    plt.xlabel('Date')
    plt.ylabel('Number of Cases')
    plt.show()
```

India: Exponential growth

```python
plot_cumulative_cases("India")
```

Poland: Between April and July, there is linear growth

```python
plot_cumulative_cases("Poland")
```

### More options

```python
bike_sharing = pd.read_csv('datasets/bike_sharing.csv')
bike_sharing.head()
```

```python
# Visualizing how the total number of bike rentals changed over time.
plt.plot(bike_sharing['dteday'], bike_sharing['cnt'])
plt.show()
```

The black lines at the bottom are related to the data type of the dteday column  the data type is object (pandas equivalent of a string). When we pass a series of strings to plt.plot(), Matplotlib doesn't know how to handle that very well. If the strings are convertible, it's better to convert them explicitly to either a numerical data type or a datetime.

```python
bike_sharing['dteday'] = pd.to_datetime(bike_sharing['dteday'])
plt.plot(bike_sharing['dteday'], bike_sharing['cnt'])
plt.show()
```

Now the dates on the bottom of the graph are overlapping. To fix this, we can rotate the labels using the rotation parameter (to control the angle of rotation) of the plt.xticks() function 

```python
plt.plot(bike_sharing['dteday'], bike_sharing['cnt'])
plt.xticks(rotation=45)
plt.title("Number of bikes rented each day")
plt.show()
```

Notice that over the 3 years, the number of rented bikes increases from Jan-May, plateus from May-Sept, then drops in the winter months. These similarities are called seasonal trends. This is probably being caused due to temperature changes over the months.<br> In time series data, specific patterns occurring regularly at specific intervals of time are called seasonality.

```python
# Look for seasonality in temperature over time
plt.plot(bike_sharing["dteday"], bike_sharing["temp"])
plt.ylabel("Temperature")
plt.xticks(rotation = 45)
plt.show()
```

We see that when the temperature grows, the number of bike rentals tends to grow too (from earlier graph). The values in the two columns follow similar patterns in how they change.

```python
# Plot the casual and registered bikes rented each day
plt.plot(bike_sharing['dteday'], bike_sharing['casual'], label = 'Casual')
plt.plot(bike_sharing['dteday'], bike_sharing['registered'], label = 'Registered')

plt.xlabel('Date')
plt.ylabel('Bikes Rented')
plt.title('Bikes Rented: Casual vs. Registered')
plt.xticks(rotation = 30)

plt.legend()
plt.show()
```

### Scatterplots
Each point (also called a marker) on the scatter plot has an x-coordinate and an y-coordinate. 
The first column we pass to plt.scatter(x, y) provides the x-coordinates, and the second one the y-coordinates.

Above, we saw that no. of bike rentals approximately increase/descrease with temperature. We visualize the relationship as a scatterplot.

```python
# Relationship between temp and cnt
plt.scatter(bike_sharing["temp"], bike_sharing["cnt"])
plt.xlabel('Temperature')
plt.ylabel('Bikes Rented')
plt.show()
```

This shows a general upward trend - **Positive correlation**

```python
# Relationship between windspeed and cnt
plt.scatter(bike_sharing['windspeed'], bike_sharing['cnt'])
plt.xlabel('Wind Speed')
plt.ylabel('Bikes Rented')
plt.show()
```

This shows an approximate downward trend - **Negative correlation**

### Correlation
- Two columns are statistically related in a specific way and to a certain degree
- Columns can be positively or negatively correlated (or not correlated at all)
- Some pairs of columns have a stronger (or weaker) correlation than others. We measure correlation strength is by calculating the degree to which the points on a scatter plot fit on a straight line.

Pearson correlation coefficient:
- Measures how well the points fit on a straight line 
- Values lie between -1.00 and +1.00.
- +1.00 means perfect positive correlation and -1.00 means perfect negative correlation. 0.0 shows no correlation.

```python
# Finding pearson's r
bike_sharing['temp'].corr(bike_sharing['cnt'])
```

```python
bike_sharing['windspeed'].corr(bike_sharing['cnt'])
```

```python
# To get pearson's r between all column pairs
bike_sharing.corr()
```

```python
# We can check the correlations of a particular column with other columns
bike_sharing.corr()['cnt'][['cnt', 'casual', 'registered']]
```

```python
print('Correlation: ', bike_sharing['temp'].corr(bike_sharing['atemp']))
plt.scatter(bike_sharing['temp'], bike_sharing['atemp'])
plt.xlabel('Air Temperature')
plt.ylabel('Feeling Temperature')
plt.show()
```

```python
print('Correlation: ', bike_sharing['windspeed'].corr(bike_sharing['hum']))
plt.scatter(bike_sharing['windspeed'], bike_sharing['hum'])
plt.xlabel('Wind Speed')
plt.ylabel('Humidity')
plt.show()
```
