# Data Visualization with Matplotlib

```python
# We generally import the pyplot submodule instead of the whole module
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Line Plot

```python
# plt.plot() generates a line plot by default
month_number = [1, 2, 3, 4, 5, 6, 7]
new_cases = [9926, 76246, 681488, 2336640,
             2835147, 4226655, 6942042]

# The first array gives the x-coordinates, and the second array gives the y-coordinates.
plt.plot(month_number, new_cases)
plt.show()
```

```python
plt.plot(month_number, new_cases)
plt.ticklabel_format(axis='y', style='plain')  # To remove scientific notation 
plt.title('New Reported Cases By Month (Globally)')  # Title of graph
plt.xlabel('Month Number')  # X-axis label
plt.ylabel('Number Of Cases')  # Y-axis label
plt.show()
```

```python
import pandas as pd
who_time_series = pd.read_csv('datasets/WHO_time_series.csv')
who_time_series.head()
```

```python
# Make Date_reported a datetime column
who_time_series["Date_reported"] = pd.to_datetime(who_time_series["Date_reported"])
```

```python
who_time_series.info()
```

```python
# Function to plot the cumulative cases for a country over the months
def plot_cumulative_cases(country_name):
    country = who_time_series[who_time_series['Country'] == country_name]
    plt.plot(country['Date_reported'], country['Cumulative_cases'])
    plt.title('{}: Cumulative Reported Cases'.format(country_name))
    plt.xlabel('Date')
    plt.ylabel('Number of Cases')
    plt.show()
```

Italy: Between March and July, there is a logarithmic growth.

```python
plot_cumulative_cases("Italy")
```

India: Exponential growth

```python
plot_cumulative_cases("India")
```

Poland: Between April and July, there is linear growth

```python
plot_cumulative_cases("Poland")
```

```python
# Plot cumulative cases for UK, Italy and France
# To know which line is which, add a label argument to the plt.plot() function, and then use the plt.legend() function

france = who_time_series[who_time_series['Country'] == 'France']
uk = who_time_series[who_time_series['Country'] == 'The United Kingdom']
italy = who_time_series[who_time_series['Country'] == 'Italy']

plt.plot(france['Date_reported'], france['Cumulative_cases'],
         label='France')
plt.plot(uk['Date_reported'], uk['Cumulative_cases'],
         label='The UK')
plt.plot(italy['Date_reported'], italy['Cumulative_cases'], 
         label = 'Italy')

plt.legend()
plt.show()
```

### More options

```python
bike_sharing = pd.read_csv('datasets/bike_sharing.csv')
bike_sharing.head()
```

```python
# Visualizing how the total number of bike rentals changed over time.
plt.plot(bike_sharing['dteday'], bike_sharing['cnt'])
plt.show()
```

The black lines at the bottom are related to the data type of the dteday column  the data type is object (pandas equivalent of a string). When we pass a series of strings to plt.plot(), Matplotlib doesn't know how to handle that very well. If the strings are convertible, it's better to convert them explicitly to either a numerical data type or a datetime.

```python
bike_sharing['dteday'] = pd.to_datetime(bike_sharing['dteday'])
plt.plot(bike_sharing['dteday'], bike_sharing['cnt'])
plt.show()
```

Now the dates on the bottom of the graph are overlapping. To fix this, we can rotate the labels using the rotation parameter (to control the angle of rotation) of the plt.xticks() function 

```python
plt.plot(bike_sharing['dteday'], bike_sharing['cnt'])
plt.xticks(rotation=45)
plt.title("Number of bikes rented each day")
plt.show()
```

Notice that over the 3 years, the number of rented bikes increases from Jan-May, plateus from May-Sept, then drops in the winter months. These similarities are called seasonal trends. This is probably being caused due to temperature changes over the months.<br> In time series data, specific patterns occurring regularly at specific intervals of time are called seasonality.

```python
# Look for seasonality in temperature over time
plt.plot(bike_sharing["dteday"], bike_sharing["temp"])
plt.ylabel("Temperature")
plt.xticks(rotation = 45)
plt.show()
```

We see that when the temperature grows, the number of bike rentals tends to grow too (from earlier graph). The values in the two columns follow similar patterns in how they change.

```python
# Plot the casual and registered bikes rented each day
plt.plot(bike_sharing['dteday'], bike_sharing['casual'], label = 'Casual')
plt.plot(bike_sharing['dteday'], bike_sharing['registered'], label = 'Registered')

plt.xlabel('Date')
plt.ylabel('Bikes Rented')
plt.title('Bikes Rented: Casual vs. Registered')
plt.xticks(rotation = 30)

plt.legend()
plt.show()
```

## Scatterplots
Each point (also called a marker) on the scatter plot has an x-coordinate and an y-coordinate. 
The first column we pass to plt.scatter(x, y) provides the x-coordinates, and the second one the y-coordinates.

Above, we saw that no. of bike rentals approximately increase/descrease with temperature. We visualize the relationship as a scatterplot.

```python
# Relationship between temp and cnt
plt.scatter(bike_sharing["temp"], bike_sharing["cnt"])
plt.xlabel('Temperature')
plt.ylabel('Bikes Rented')
plt.show()
```

This shows a general upward trend - **Positive correlation**

```python
# Relationship between windspeed and cnt
plt.scatter(bike_sharing['windspeed'], bike_sharing['cnt'])
plt.xlabel('Wind Speed')
plt.ylabel('Bikes Rented')
plt.show()
```

This shows an approximate downward trend - **Negative correlation**

### Correlation
- Two columns are statistically related in a specific way and to a certain degree
- Columns can be positively or negatively correlated (or not correlated at all)
- Some pairs of columns have a stronger (or weaker) correlation than others. We measure correlation strength is by calculating the degree to which the points on a scatter plot fit on a straight line.

Pearson correlation coefficient:
- Measures how well the points fit on a straight line 
- Values lie between -1.00 and +1.00.
- +1.00 means perfect positive correlation and -1.00 means perfect negative correlation. 0.0 shows no correlation.

```python
# Finding pearson's r
bike_sharing['temp'].corr(bike_sharing['cnt'])
```

```python
bike_sharing['windspeed'].corr(bike_sharing['cnt'])
```

```python
# To get pearson's r between all column pairs
bike_sharing.corr()
```

```python
# We can check the correlations of a particular column with other columns
bike_sharing.corr()['cnt'][['cnt', 'casual', 'registered']]
```

```python
print('Correlation: ', bike_sharing['temp'].corr(bike_sharing['atemp']))
plt.scatter(bike_sharing['temp'], bike_sharing['atemp'])
plt.xlabel('Air Temperature')
plt.ylabel('Feeling Temperature')
plt.show()
```

```python
print('Correlation: ', bike_sharing['windspeed'].corr(bike_sharing['hum']))
plt.scatter(bike_sharing['windspeed'], bike_sharing['hum'])
plt.xlabel('Wind Speed')
plt.ylabel('Humidity')
plt.show()
```

For categorical data, we can encode it in numbers to calculate correlation. <br>
Ex: workingday column is encoded (1 means a working day and 0 means a non-working day)

```python
bike_sharing["workingday"].value_counts()
```

```python
# Calculate its correlation with casual and registered users
bike_sharing.corr()['workingday'][['casual', 'registered']]
```

- We see a negative correlation with casual column and a positive correlation with registered column.
- This means that registered users tend to use the bikes more on working days, while casual (non-registered) users tend to rent bikes more on the weekends and holidays.

However, the 1 and 0 encoding is arbitrary. We can reassign them:

```python
bike_sharing['workingday'].replace({1:0, 0:1}, inplace=True)
```

```python
bike_sharing.corr()['workingday'][['casual', 'registered']]
```

The correlation strengths remained the same, but their sign changed. Because the encoding is arbitrary, the correlation can be both positive and negative. <br>
Taking another example, weathersit column (encoded from 1-4). The higher the value, the worse the weather.

```python
# Checking its correlation with count and humidity
bike_sharing.corr()['weathersit'][['cnt', 'hum']]
```

- We see a negative correlation with count column. This means that as the weather gets worse (value increases), the count of rented bikes decreases.
- We see a positive correlation with humidity. This means that as the weather gets worse, humidity increases.

Note: Correlation does not always mean causality.

```python
# Visualizing categorical correlations as scatterplots
plt.scatter(bike_sharing["workingday"], bike_sharing["casual"])
plt.title('Working Day Vs. Casual')
plt.show()

plt.scatter(bike_sharing["workingday"], bike_sharing["registered"])
plt.title('Working Day Vs. Registered')
plt.show()
```

```python
# Calculating the average number of rentals we see for casual and registered users on working and non-working days
# Note that the encoding was reversed

casual_nw = bike_sharing.loc[bike_sharing["workingday"] == 1, "casual"].mean()
casual_w = bike_sharing.loc[bike_sharing["workingday"] == 0, "casual"].mean()
registered_nw = bike_sharing.loc[bike_sharing["workingday"] == 1, "registered"].mean()
registered_w = bike_sharing.loc[bike_sharing["workingday"] == 0, "registered"].mean()

summary = {
    'casual' : [casual_nw, casual_w],
    'registered' : [registered_nw, registered_w]
}
working_days_users = pd.DataFrame(summary, index = ['Non-working (1)', 'Working (0)'])
working_days_users
```

This corroborates our hypothesis made from the correlations.

## Histograms
Each bar represents an interval, not a single value.

Grouped frequency table:
- For numerical data that is continuous, we create a grouped frequency table because there are too many unique values.
- We group the unique values into equal intervals.
- They can be visualized using histograms.

```python
# Grouped frequency table for registered users
# bins = no. of intervals
bike_sharing['registered'].value_counts(bins=10).sort_index()
```

```python
# Visualizing using histogram
# plt.hist() function automatically generated a grouped frequency table with ten equal intervals for the cnt column.
# Then it plotted a bar for each intervalwhere the height of each bar corresponds to the frequency of that interval.

plt.hist(bike_sharing['registered'])
plt.show()
```

This distribution resembles a **Normal or Gaussian distribution.**
- Most of the values in that column lie in the middle of the range => On most days, registered users rent between 3000 and 4000 bikes.
- As we approach the minimum and the maximum points of the range, we have less and less of those values => Registered users renting fewer than 1000 bikes a day is rare.

```python
# Now, we analyze the same for casual users
plt.hist(bike_sharing['casual'])
plt.show()
```

This distribution resembles a **Skewed distribution.**
- The values pile up toward the end or the starting point of the range, making up the body of the distribution.
- Then the values decrease in frequency towards the opposite end, forming the tail of the distribution.
- This is a Right-skewed distribution.

## Example: Traffic Analysis
Try to find the causes for bad traffic.

```python
traffic = pd.read_csv("datasets/traffic_sao_paulo.csv", sep=';')
traffic.head()
```

```python
# Converting slowness% from string to float
traffic['Slowness in traffic (%)'] = traffic['Slowness in traffic (%)'].str.replace(',', '.')
traffic['Slowness in traffic (%)'] = traffic['Slowness in traffic (%)'].astype(float)
```

```python
# Check the distribution of values in slowness column

traffic['Slowness in traffic (%)'].value_counts(bins=10).sort_index()
```

```python
# Plotting a histogram
# plt.hist(traffic['Slowness in traffic (%)'])
# plt.show()

# Alternate way to plot using pandas directly
# Pandas uses matplotlib under the hood
traffic['Slowness in traffic (%)'].plot.hist()
plt.title("Distribution of Slowness in traffic (%)")
plt.xlabel("Slowness in traffic (%)")
plt.show()
```

```python
# Now we analyze all the 'incident' related columns
# The value shows the number of incidents

incidents = traffic.drop(['Hour (Coded)', 'Slowness in traffic (%)'],
                        axis=1)
incidents.head()
```

```python
# Total incidents of each type
incidents.sum()
```

```python
# Compare them by plotting a bar graph
incidents.sum().plot.barh()
plt.show()
```

```python
# To find the relationship of each incident type with traffic slowness, we can look up correlation values
traffic.corr()['Slowness in traffic (%)']
```

Among the incident types, Slowness in traffic (%) correlates most strongly with the following:
- Lack of electricity: +0.44
- Point of flooding: +0.42
- Semaphore off: +0.34

```python
# Visualize correlation with a scatterplot

# plt.scatter(traffic['Slowness in traffic (%)'], traffic['Lack of electricity'])
# plt.show()

# Alternate way: Generate using pandas directly
traffic.plot.scatter(x='Slowness in traffic (%)', y='Lack of electricity')
plt.show()

traffic.plot.scatter(x='Slowness in traffic (%)', y='Point of flooding')
plt.show()

traffic.plot.scatter(x='Slowness in traffic (%)', y='Semaphore off')
plt.show()
```

We further analyze the incident frequencies on the rows that have slowness >= 20%.

```python
slowness_20_or_more = traffic[traffic["Slowness in traffic (%)"] >= 20]
slowness_20_or_more = slowness_20_or_more.drop(["Hour (Coded)", "Slowness in traffic (%)"], axis=1)
incident_frequencies = slowness_20_or_more.sum()
incident_frequencies
```

```python
# Plotting them as a bar graph
incident_frequencies.plot.barh()
plt.show()
```

We see that lack of electricity and flooding are two factors that are causing significant traffic problems.

<br>Next, we isolate the data for each day of the week.

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
traffic_per_day = {}
for i, day in zip(range(0, 135, 27), days):
    each_day_traffic = traffic[i:i+27]
    traffic_per_day[day] = each_day_traffic

# The key is the day of the week, the value is a dataframe of data for that day
traffic_per_day["Monday"].head(3)
```

```python
# To see how to values in a column change over time, we can use a line plot
# Here, we see how the slowness changes over the hours for each day of the week

for day in days:
    traffic_per_day[day].plot.line(x='Hour (Coded)',
                                   y='Slowness in traffic (%)')
    plt.title(day)
    plt.ylim([0, 25]) # Sets the range of the y-axis
    plt.show()
```

```python
# To make comparison easier, we can plot them all in the same plot
# cannot use DataFrame.plot.line() here
for day in days:
    plt.plot(traffic_per_day[day]['Hour (Coded)'],
             traffic_per_day[day]['Slowness in traffic (%)'], 
            label = day)

plt.legend()
plt.show()
```

## Grid Chart
-  Collection of similar graphs that usually share the same x- and y-axis range to ease comparison.
- Also called small multiple.
- It has table-like structure with rows and columns. These can be specified using plt.subplot(nrows, ncols, index) where index is the position of the plot on the grid.
- We start by creating a figure using plt.figure() that will contain the subplots.
- After creating each subplot, we plot the actual graph using plt.plot() or other methods.

<br> We represent the plots above in a grid chart.

```python
#We can increase the size of the entire figure by using plt.figure(figsize=(width, height))

plt.figure(figsize=(10,12))

# For each day of the week
for i, day in zip(range(1,6), days):
    plt.subplot(3,2,i)
    plt.plot(traffic_per_day[day]["Hour (Coded)"],
             traffic_per_day[day]["Slowness in traffic (%)"])
    plt.title(day)
    plt.ylim(0,25)
     
# For the cumulative graph
plt.subplot(3,2,6)
for day in days:
    plt.plot(traffic_per_day[day]['Hour (Coded)'],
             traffic_per_day[day]['Slowness in traffic (%)'], 
            label = day)
plt.ylim([0,25])
plt.legend()

plt.show()
```

## Enhancing graphs using OO interface

```python
# Using functional interface

# Plotting bar graph for slowness % every hour
plt.barh(traffic["Hour (Coded)"], traffic["Slowness in traffic (%)"])
plt.show()
```

To create a graph using the OO interface, we use the plt.subplots() function, which generates an empty plot and returns two objects: <br>
1. The Figure (the canvas)
2. The Axes (the plot)

```python
fig, ax = plt.subplots()
ax.barh(traffic["Hour (Coded)"], traffic["Slowness in traffic (%)"])
plt.ylabel("Hour")
plt.show()
```

```python
# To make a graph mobile friendly, we can set the dimensions using figsize
fig, ax = plt.subplots(figsize=(5,7.5))
ax.barh(traffic["Hour (Coded)"], traffic["Slowness in traffic (%)"])
plt.ylabel("Hour")
plt.show()
```

```python
# Removing the axes and ticks
fig, ax = plt.subplots()
ax.bar(traffic["Hour (Coded)"], traffic["Slowness in traffic (%)"])
plt.ylabel("Hour")

for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)
ax.tick_params(bottom=False, left=False)

plt.show()
```

```python
# We can set only selected ticks
fig, ax = plt.subplots()
ax.bar(traffic["Hour (Coded)"], traffic["Slowness in traffic (%)"])
plt.xlabel("Slowness")
plt.ylabel("Hour")

for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)
ax.set_xticks([0, 10, 20, 30])

plt.show()
```

```python
# We can set the bar thickness using 
# height parameter in a horizontal plot
# or width parameter in a vertical plot

fig, ax = plt.subplots()
ax.barh(traffic["Hour (Coded)"], traffic["Slowness in traffic (%)"],
       height = 0.1)
plt.ylabel("Hour")
plt.show()
```

```python
# We can move x-axis ticks to the top to make the graph easier to read
# We also change their colour to make them less distracting

fig, ax = plt.subplots()
ax.barh(traffic["Hour (Coded)"], traffic["Slowness in traffic (%)"])
plt.ylabel("Hour")

for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)

ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')
ax.tick_params(axis='x', colors='grey')
ax.tick_params(top=False, left=False)

plt.show()
```

```python
# We can change colour of the bars
fig, ax = plt.subplots()
ax.barh(traffic["Hour (Coded)"], traffic["Slowness in traffic (%)"],
       color='#af0b1e')
plt.ylabel("Hour")

plt.show()
```

```python
# We can add a title and subtitle

fig, ax = plt.subplots()
ax.barh(traffic["Hour (Coded)"], traffic["Slowness in traffic (%)"],)
plt.ylabel("Hour")

ax.text(x=8, y=32, s='Slowness (%)',
       size=15, weight='bold')
ax.text(x=10, y=30, s='in traffic',
       size=12)

plt.show()
```

```python
# We can add lines to the graph to make it easier to read
fig, ax = plt.subplots()
ax.barh(traffic["Hour (Coded)"], traffic["Slowness in traffic (%)"],)
plt.ylabel("Hour")
plt.xlim(0,25)

ax.axvline(x=10, c='grey', alpha=0.5)
plt.show()
```

```python
# Now we put it all together to create the final graph

fig, ax = plt.subplots(figsize=(7,7))
ax.barh(traffic["Hour (Coded)"], traffic["Slowness in traffic (%)"],
       height = 0.7, color='#af0b1e')
plt.ylabel("Hour", size=15)

for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)
    
ax.set_xticks([0, 10, 20, 30])
ax.xaxis.tick_top()
ax.tick_params(axis='x', colors='grey')
ax.tick_params(top=False, left=False)

ax.text(x=8, y=32, s='Slowness (%)',
       size=15, weight='bold')
ax.text(x=10, y=31, s='in traffic',
       size=12)

ax.axvline(x=10, c='grey', alpha=0.5)

plt.show()
```
