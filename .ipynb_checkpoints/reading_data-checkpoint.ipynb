```python
# copy data from dict data to list

with open('albumlist.csv', 'r') as my_file:
    reader = csv.DictReader(my_file)
    
    albums = []
    for row in reader:
        albums.append(row)
        
print("Number of albums: ", len(albums))

#albums - list of dictionaries
```

```python
#Get albums from year 1974
albums_1974 = []
for row in albums:
    if row['Year'] == '1974':
        albums_1974.append(row)
        
print(len(albums_1974))
        
for album in albums_1974:
    print(album)
    
#Using list comprehension

albums_1974 = [row for row in albums if row['Year'] == '1974']
```

```python
for album in albums_1974:
    print(album['Album'], 'by', album['Artist'])
```

```python
#Get 3 albums from rock
ls = []
for row in albums:
    if album['Genre'] == 'Rock':
        ls.append(row['Album'])
print(ls[:3]) 

print('\n')
#Using list comprehension

print([row['Album'] for row in albums if row["Genre"] == "Rock"][:3])
```

```python
hiphop_count = 0
for album in albums:
    if album["Genre"] == "Hip Hop":
        hiphop_count += 1
print("No. of hiphop albums: ", hiphop_count)
```

```python
rock_albums = [row for row in albums if(row["Genre"] == "Rock" 
                and "Country Rock" in row["Subgenre"])]
for album in rock_albums:
    print(album['Album'], album['Genre'], album['Subgenre'])
```

```python
def is_valid_year(string):
    try:
        year = int(string)
    except ValueError: #Some year values may be non-numeric
        return False
    else:
        return year
    
# Add int(year) to list 
release_years = [int(row["Year"]) for row in albums if is_valid_year(row["Year"])]

print(release_years[:10])

print('Earliest release: ', min(release_years))
print('Latest release: ', max(release_years))
```

```python
latest_release_years = [int(row["Year"]) for row in albums if is_valid_year(row["Year"]) and int(row["Year"]) > 2000]
print("No of 2000s albums:", len(latest_release_years))
```

```python
#Sorting

def album_length(album):
    return len(album["Album"])

def album_year(album):
    return is_valid_year(album["Year"])

name_sorted = sorted(albums, key = album_length) # by album name length
year_sorted = sorted(albums, key = album_year)  #by album year

#Using lambda functions

year_sorted_l = sorted(albums, key = lambda album: is_valid_year(album["Year"]))  #by album year
artist_sorted_l = sorted(albums, key = lambda album: album["Artist"]) #by artist name

artist_sorted_l
```
