# Data Visualization with Seaborn

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing = pd.read_csv("datasets/house_data.csv")
housing.head(3)
```

With seaborn, we can represent more than two variables on the graph.<br>
We start by plotting 2 columns, then add more. <br>
<br>
Relational plot (shows relationships between variables) is created using relplot() and it creates a scatterplot by default.

```python
# switches to seaborn defaults instead of matplotlib defaults
# All subsequent graphs inherit seaborn style
sns.set_theme()
```

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice')
plt.show()
```

Now we add another variables using the hue parameter.

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual')
plt.show()
```

```python
# We can choose the colour palette
# Find all palettes at: https://matplotlib.org/tutorials/colors/colormaps.html

sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn')
plt.show()
```

Now, we add another variable using size parameter.
<br> To make the size difference noticeable, use the sizes parameter. 

```python
# Using marker size for a numerical variable (Garage Area)
# We pass in a tuple with minimum and maximum size.

sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300))
plt.show()
```

```python
# Using marker size for a categorical variable (Rooms)
# We pass in a list of values for each category (here there are 2 categories)

sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Rooms', sizes=[200, 50])
plt.show()
```

We can add yet another variable using the shape of the marker specified by the 'style' parameter.
<br> The 'markers' parameter is used to specify the shapes.

```python
# Using marker shape to add Rooms variable
# More marker shape options: https://matplotlib.org/3.3.3/api/markers_api.html

sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300),
            style='Rooms',  markers=['*','v'])
plt.show()
```

To add another variable, we can spatially split the graph. For each value of this variable, we'll have a new plot with the other 5 variables.<br>
We use the 'col' parameter for this.

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300),
            style='Rooms', col='Year')
plt.show()
```
