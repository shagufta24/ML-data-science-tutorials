```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv', header = [0, 1])
df.head()
```

```python
df.loc[: , ('Unnamed: 1_level_0', ['Player'])] = df.loc[: , ('Unnamed: 1_level_0', ['Player'])].replace(r".*\\", "", regex=True)
```

```python
pos1 = df['Unnamed: 3_level_0']['Pos'].str.extract(r'(..)', expand=True)
pos2 = df['Unnamed: 3_level_0']['Pos'].str.extract(r'..(.*)', expand=True)
temp = pd.concat([pos1, pos2], axis = 1)
temp.columns = pd.MultiIndex.from_tuples([('Pos','Pos1'),('Pos','Pos2')])
df = df.drop(df.iloc[:, [3]], axis = 1)
df = pd.concat([df, temp], axis = 1)
```
