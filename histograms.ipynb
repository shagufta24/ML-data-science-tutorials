 # Histograms
 - Probability distribution of a continuous variable.
 - The x-axis of the histogram denotes the number of bins while the y-axis represents the frequency of a particular bin. 

### Generating data

```python
import numpy as np

np.random.seed(100)
np_hist = np.random.normal(loc=0.0, scale=1, size=1000) # Draw random samples from a normal/gaussian distribution
```

```python
np_hist.mean(), np_hist.std()
```

```python
# hist (frequency) will be the y-axis
# bin_edges will be the x-axis

hist, bin_edges = np.histogram(np_hist)
hist, bin_edges
```

```python
# Rounding x-axis values to display better
bin_edges = np.round(bin_edges,0)
```

### Plotting the histogram using matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create new figure
plt.figure(figsize=[10,8])

# Make bar plot
plt.bar(x = bin_edges[:-1], height = hist, width = 0.5)

# X-axis limits
plt.xlim(min(bin_edges), max(bin_edges))

# Display grid lines
plt.grid(axis='both', alpha=0.75)

# Set axis labels
plt.xlabel('Value',fontsize=15)
plt.ylabel('Frequency',fontsize=15)

# Set axis labels size
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.title('Normal Distribution Histogram',fontsize=15)
plt.show()
```

### Using only matplotlib

```python
plt.figure(figsize=[10,8])

# Make histogram directly from the data
n, bins, patches = plt.hist(x=np_hist, bins='auto', rwidth=0.85)

plt.grid(axis='y', alpha=0.75)
plt.xlabel('Value',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.title('Normal Distribution Histogram',fontsize=15)
plt.show()
```
